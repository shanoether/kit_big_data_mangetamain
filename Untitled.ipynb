```python
import polars as pl
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# 1. Caractéristiques de base
user_features = (
    df_interaction
    .group_by("user_id")
    .agg([
        pl.col("rating").mean().alias("avg_rating"),
        pl.col("rating").std().alias("std_rating"),
        pl.len().alias("n_reviews"),
    ])
)

# 2. TF-IDF
tfidf = TfidfVectorizer(max_features=50, stop_words="english")
tfidf_matrix = tfidf.fit_transform(df_interaction["review"].fill_null(""))
user_tfidf = (
    pl.DataFrame(tfidf_matrix.toarray(), schema=[f"tfidf_{i}" for i in range(50)])
    .with_columns(pl.Series("user_id", df_interaction["user_id"]))
    .group_by("user_id")
    .agg([pl.col(f"tfidf_{i}").mean() for i in range(50)])
)

# 3. Fusion des données
user_matrix = user_features.join(user_tfidf, on="user_id", how="left")
user_matrix = user_matrix.fill_null(0)

# 4. Normalisation et k-means
X = StandardScaler().fit_transform(user_matrix.drop("user_id").to_pandas())
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
user_matrix = user_matrix.with_columns(pl.Series("cluster", kmeans.labels_))

# 5. Résultats agrégés
cluster_summary = (
    user_matrix
    .group_by("cluster")
    .agg([
        pl.col("avg_rating").mean().alias("moyenne_notes"),
        pl.col("std_rating").mean().alias("ecart_type_notes"),
        pl.col("n_reviews").mean().alias("moyenne_reviews"),
        pl.len().alias("nombre_utilisateurs"),
    ])
)
print("\nRésumé des clusters :")
print(cluster_summary)

# 6. Top 10 mots par cluster (version corrigée)
for cluster in range(k):
    # Filtrer les utilisateurs du cluster
    cluster_data = user_matrix.filter(pl.col("cluster") == cluster)

    # Extraire les colonnes TF-IDF et convertir en numpy array
    tfidf_columns = [f"tfidf_{i}" for i in range(50)]
    cluster_tfidf = cluster_data.select(tfidf_columns).to_numpy()  # Conversion en numpy array

    # Calculer la moyenne des scores TF-IDF pour ce cluster
    mean_tfidf_scores = np.mean(cluster_tfidf, axis=0)  # Moyenne sur l'axe des utilisateurs

    # Trouver les indices des 10 mots avec les scores les plus élevés
    top_terms_idx = np.argsort(-mean_tfidf_scores)[:10]  # Indices triés par score décroissant

    # Récupérer les noms des termes (sans utiliser Polars pour éviter les conflits de types)
    top_terms = [tfidf.get_feature_names_out()[i] for i in top_terms_idx]

    print(f"\nCluster {cluster} - Top 10 mots les plus utilisés :")
    for i, term in enumerate(top_terms, 1):
        score = mean_tfidf_scores[top_terms_idx[i-1]]  # Score moyen pour ce terme
        print(f"{i}. {term} (score TF-IDF moyen : {score:.4f})")

# 7. Visualisation PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap="viridis", alpha=0.6)
plt.xlabel("Composante principale 1")
plt.ylabel("Composante principale 2")
plt.title(f"Clusters d'utilisateurs (k={k})")
plt.colorbar(scatter, label="Cluster")
plt.grid()
plt.show()
```

```python
import polars as pl
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Convertir la colonne 'date' en datetime
df_interaction = df_interaction.with_columns(
    pl.col("date").str.strptime(pl.Date, "%Y-%m-%d")
)

# Extraire des caractéristiques temporelles par utilisateur
user_time_features = (
    df_interaction
    .group_by("user_id")
    .agg([
        # Nombre total de reviews
        pl.len().alias("n_reviews"),
        # Date de la première review
        pl.col("date").min().alias("first_review"),
        # Date de la dernière review
        pl.col("date").max().alias("last_review"),
        # Durée d'activité (en jours)
        (pl.col("date").max() - pl.col("date").min()).dt.days().alias("activity_duration"),
        # Nombre moyen de reviews par mois
        (pl.len() / ((pl.col("date").max() - pl.col("date").min()).dt.days() / 30)).alias("reviews_per_month"),
        # Jour de la semaine le plus fréquent (0=lundi, 6=dimanche)
        pl.col("date").dt.weekday().mode().first().alias("favorite_weekday"),
        # Mois de l'année le plus fréquent (1-12)
        pl.col("date").dt.month().mode().first().alias("favorite_month"),
        # Heure de la journée la plus fréquente (si disponible)
        # pl.col("date").dt.hour().mode().first().alias("favorite_hour"),
    ])
)

# Ajouter des caractéristiques basées sur la distribution des dates
def get_time_features(df):
    dates = df["date"].to_list()
    if len(dates) < 2:
        return {
            "review_interval_mean": 0,
            "review_interval_std": 0,
            "activity_regularity": 0,
        }
    intervals = np.diff(sorted(dates))
    return {
        "review_interval_mean": np.mean(intervals),
        "review_interval_std": np.std(intervals),
        "activity_regularity": 1 / (1 + np.std(intervals)),  # 1 = très régulier
    }

user_time_features = user_time_features.with_columns(
    pl.struct(["date"]).map_elements(
        lambda x: get_time_features(x),
        return_dtype=pl.Struct([
            pl.Field("review_interval_mean", pl.Float64),
            pl.Field("review_interval_std", pl.Float64),
            pl.Field("activity_regularity", pl.Float64),
        ])
    ).alias("time_stats")
).unnest("time_stats")

# Remplacer les valeurs infinies/NaN
user_time_features = user_time_features.fill_null(0)
user_time_features = user_time_features.fill_nan(0)
```

```python
# Encoder les jours de la semaine et mois de manière cyclique
user_time_features = user_time_features.with_columns([
    # Jour de la semaine (sin/cos)
    pl.col("favorite_weekday").cast(pl.Float64) / 7 * 2 * np.pi,
    (pl.col("favorite_weekday").cast(pl.Float64) / 7 * 2 * np.pi).sin().alias("weekday_sin"),
    (pl.col("favorite_weekday").cast(pl.Float64) / 7 * 2 * np.pi).cos().alias("weekday_cos"),
    # Mois de l'année (sin/cos)
    pl.col("favorite_month").cast(pl.Float64) / 12 * 2 * np.pi,
    (pl.col("favorite_month").cast(pl.Float64) / 12 * 2 * np.pi).sin().alias("month_sin"),
    (pl.col("favorite_month").cast(pl.Float64) / 12 * 2 * np.pi).cos().alias("month_cos"),
])
```

```python
# Sélectionner les colonnes pour le clustering
time_columns = [
    "n_reviews", "activity_duration", "reviews_per_month",
    "review_interval_mean", "review_interval_std", "activity_regularity",
    "weekday_sin", "weekday_cos", "month_sin", "month_cos"
]

# Normaliser les données
scaler = StandardScaler()
X = scaler.fit_transform(user_time_features.select(time_columns).to_pandas())

# Méthode du coude et silhouette pour choisir k
inertia = []
silhouette_scores = []
k_range = range(2, 8)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Afficher les graphiques pour choisir k
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.xlabel("Nombre de clusters (k)")
plt.ylabel("Inertie")
plt.title("Méthode du coude")

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel("Nombre de clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score")
plt.grid()
plt.tight_layout()
plt.show()

# Appliquer k-means avec le k optimal (exemple: k=4)
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
user_time_features = user_time_features.with_columns(
    pl.Series("cluster", kmeans.labels_)
)

# Silhouette score final
silhouette_avg = silhouette_score(X, kmeans.labels_)
print(f"\nSilhouette Score pour k={k}: {silhouette_avg:.4f}")
```

```python
from sklearn.metrics import silhouette_samples

sample_silhouette_values = silhouette_samples(X, kmeans.labels_)
plt.figure(figsize=(10, 6))
y_lower = 10
for i in range(k):
    ith_cluster_silhouette_values = sample_silhouette_values[kmeans.labels_ == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0, ith_cluster_silhouette_values,
        facecolor=plt.cm.viridis(float(i) / k),
        edgecolor=plt.cm.viridis(float(i) / k),
        alpha=0.7,
    )
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

plt.title(f"Graphique des silhouettes pour k={k}")
plt.xlabel("Silhouette coefficient")
plt.ylabel("Cluster")
plt.axvline(x=silhouette_avg, color="red", linestyle="--")
plt.xlim([-0.1, 1])
plt.grid()
plt.show()
```

```python
import polars as pl
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans  # ⚡ plus rapide que KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# 1. Conversion des dates et timestamps
df_interaction = df_interaction.with_columns(
    pl.col("date").cast(pl.Utf8).str.strptime(pl.Datetime, "%Y-%m-%d")
).with_columns(
    pl.col("date").dt.timestamp("ms").alias("timestamp")
)

# 2. Caractéristiques temporelles de base
user_time_features = (
    df_interaction
    .group_by("user_id")
    .agg([
        pl.len().alias("n_reviews"),
        pl.col("timestamp").min().alias("first_review"),
        pl.col("timestamp").max().alias("last_review"),
    ])
    .with_columns(
        ((pl.col("last_review") - pl.col("first_review")) / (1000 * 60 * 60 * 24)).alias("activity_duration")
    )
    .with_columns(
        (pl.col("n_reviews") / ((pl.col("activity_duration") + 1) / 30)).alias("reviews_per_month")
    )
)

# 3. Extraire les jours de la semaine et mois directement avec Polars
df_interaction = df_interaction.with_columns([
    pl.col("date").dt.weekday().alias("weekday"),
    pl.col("date").dt.month().alias("month"),
])

# 4. Calcul des intervalles entre reviews
df_sorted = df_interaction.sort(["user_id", "date"])
df_sorted = df_sorted.with_columns(
    (pl.col("date").diff().dt.total_days().over("user_id")).alias("interval_days")
)

# 5. Agrégation vectorisée des features temporelles
user_time_stats = (
    df_sorted.group_by("user_id")
    .agg([
        pl.col("weekday").mode().alias("favorite_weekday"),
        pl.col("month").mode().alias("favorite_month"),
        pl.col("interval_days").mean().fill_null(0).alias("review_interval_mean"),
        pl.col("interval_days").std().fill_null(0).alias("review_interval_std"),
    ])
    .with_columns(
        (1 / (1 + pl.col("review_interval_std"))).alias("activity_regularity")
    )
)

# 6. Fusionner les statistiques
user_time_features = user_time_features.join(user_time_stats, on="user_id", how="left")

# 7. Encodage cyclique (jours / mois)
user_time_features = user_time_features.with_columns([
    (pl.col("favorite_weekday").cast(pl.Float64) / 7 * 2 * np.pi).sin().alias("weekday_sin"),
    (pl.col("favorite_weekday").cast(pl.Float64) / 7 * 2 * np.pi).cos().alias("weekday_cos"),
    (pl.col("favorite_month").cast(pl.Float64) / 12 * 2 * np.pi).sin().alias("month_sin"),
    (pl.col("favorite_month").cast(pl.Float64) / 12 * 2 * np.pi).cos().alias("month_cos"),
])

# 8. Nettoyage des NaN
user_time_features = user_time_features.with_columns([
    pl.col("reviews_per_month").fill_nan(0),
    pl.col("review_interval_mean").fill_nan(0),
    pl.col("review_interval_std").fill_nan(0),
    pl.col("activity_regularity").fill_nan(0),
])

# 9. Sélection des colonnes de clustering
time_columns = [
    "n_reviews", "activity_duration", "reviews_per_month",
    "review_interval_mean", "review_interval_std", "activity_regularity",
    "weekday_sin", "weekday_cos", "month_sin", "month_cos"
]

# 10. Normalisation
scaler = StandardScaler()
X = scaler.fit_transform(user_time_features.select(time_columns).to_numpy())

# 11. Recherche rapide du k optimal (2 à 5)
best_k, best_score = 0, -1
for k in range(2, 6):
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=1024)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    if score > best_score:
        best_k, best_score = k, score

print(f"Meilleur k trouvé = {best_k}, silhouette = {best_score:.4f}")

# 12. Application du clustering final
kmeans = MiniBatchKMeans(n_clusters=best_k, random_state=42, batch_size=1024)
labels = kmeans.fit_predict(X)
user_time_features = user_time_features.with_columns(pl.Series("cluster", labels))

# 13. Résumé des clusters
cluster_summary = (
    user_time_features
    .group_by("cluster")
    .agg([
        pl.col("n_reviews").mean().alias("moyenne_reviews"),
        pl.col("activity_duration").mean().alias("duree_activite_moyenne"),
        pl.col("reviews_per_month").mean().alias("reviews_par_mois"),
        pl.col("review_interval_mean").mean().alias("intervalle_moyen_jours"),
        pl.col("review_interval_std").mean().alias("ecart_type_intervalle"),
        pl.col("activity_regularity").mean().alias("regularite"),
        pl.len().alias("nombre_utilisateurs"),
    ])
)
print("\nRésumé des clusters temporels :")
print(cluster_summary)

# 14. Visualisation PCA (échantillon si trop de points)
if len(X) > 5000:
    idx = np.random.choice(len(X), size=5000, replace=False)
    X_vis = X[idx]
    labels_vis = labels[idx]
else:
    X_vis = X
    labels_vis = labels

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_vis)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_vis, cmap='viridis', s=40)
plt.title(f"Clusters utilisateurs (PCA, k={best_k})")
plt.xlabel("Composante principale 1")
plt.ylabel("Composante principale 2")
plt.colorbar(label='Cluster')
plt.tight_layout()
plt.show()
```

```python
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans

# --- 1. Préparation des données ---
df = df_interaction.with_columns(
    pl.col("date").str.strptime(pl.Date, "%Y-%m-%d")
)

# Extraire année et mois
df = df.with_columns([
    pl.col("date").dt.year().alias("year"),
    pl.col("date").dt.month().alias("month")
])

# --- 2. Caractéristiques par utilisateur ---
user_features = (
    df.group_by("user_id")
    .agg([
        pl.len().alias("n_reviews"),
        pl.col("rating").mean().alias("avg_rating"),
        pl.col("recipe_id").n_unique().alias("unique_recipes"),
        pl.col("year").min().alias("first_year"),
        pl.col("year").max().alias("last_year")
    ])
)

# Durée d’activité
user_features = user_features.with_columns(
    (pl.col("last_year") - pl.col("first_year")).alias("activity_years")
)

# --- 3. Normalisation et clustering ---
cols = ["n_reviews", "avg_rating", "unique_recipes", "activity_years"]
X = user_features.select(cols).to_numpy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = MiniBatchKMeans(n_clusters=6, random_state=42, batch_size=512)
labels = kmeans.fit_predict(X_scaled)
user_features = user_features.with_columns(pl.Series("cluster", labels))

# --- 4. Joindre les clusters avec les dates originales ---
df_clustered = df.join(user_features.select(["user_id", "cluster"]), on="user_id", how="left")

# --- 5. Regrouper par mois pour voir l'évolution ---
df_time = (
    df_clustered.group_by(["year", "month", "cluster"])
    .agg(pl.col("user_id").n_unique().alias("n_users"))
    .sort(["year", "month", "cluster"])
)

# Créer une colonne datetime pour l’axe x
df_time = df_time.with_columns(
    (pl.datetime(df_time["year"], df_time["month"], 1)).alias("period")
)

# --- 6. Conversion en pandas pour matplotlib ---
df_pandas = df_time.to_pandas()

# --- 7. Visualisation temporelle ---
plt.figure(figsize=(12, 6))

for c in sorted(df_pandas["cluster"].unique()):
    df_c = df_pandas[df_pandas["cluster"] == c]
    plt.plot(df_c["period"], df_c["n_users"], marker="", label=f"Cluster {c}")

plt.title("Évolution du nombre d’utilisateurs par cluster (KMeans)")
plt.xlabel("Date")
plt.ylabel("Nombre d’utilisateurs")
plt.legend(title="Cluster")
plt.grid(True)
plt.tight_layout()
plt.show()
```
